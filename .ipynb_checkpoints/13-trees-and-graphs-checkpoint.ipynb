### Printing/Searching in a tree, and Depth First Search (DFS)

![](trees-and-graphs/Whiteboard[9]-01.png) 

```python
class Node:
    def __init__(self, val):
        self.value = val
        self.left = None
        self.right = None
    
    def __str__(self):
        return str(self.value)
```

```python
class BST:
    def __init__(self):
        self.root = None
        
    def _insert(self, start_node, new_node):
        # helper method
        if start_node.value == new_node.value:
            return None
        elif start_node.value > new_node.value:
            if start_node.left == None:
                start_node.left = new_node
            else:
                self._insert(start_node.left, new_node)
        else:
            if start_node.right == None:
                start_node.right = new_node
            else:
                self._insert(start_node.right, new_node)
    
    def insert(self, val):
        new_node = Node(val)
        if self.root == None:
            self.root = new_node
        else:
            self._insert(self.root, new_node)
    
    def printBF(self):
        q = []
        if self.root:
            q.append(self.root)
        # run until the queue is not empty
        while len(q) > 0:
            # keep adding the L and R children if they exist for the first node in the queue
            if q[0].left:
                q.append(q[0].left)
            if q[0].right:
                q.append(q[0].right)
            temp = q.pop(0)
            print(temp, end=" ")
        print()
    
    def _printInDF(self, start):
        if start == None:
            return
        self._printInDF(start.left)
        print(start, end=" ")
        self._printInDF(start.right)
        
    def _printPreDF(self, start):
        if start == None:
            return
        print(start, end=" ")
        self._printPreDF(start.left)
        self._printPreDF(start.right)
    
    def _printPostDF(self, start):
        if start == None:
            return
        self._printPostDF(start.left)
        self._printPostDF(start.right)
        print(start, end=" ")
    
    def printDF(self, order="in"):
        if order == "in":
            self._printInDF(self.root)
        elif order == "pre":
            self._printPreDF(self.root)
        else:
            self._printPostDF(self.root)
        print()
    
    
    def search(self, target):
        if self.root:
            return self._search(self.root, target)
    
    def _search(self, start, target):
        if start.value == target.value:
            return start
        elif start.value > target.value:
            # left sub-tree
            if start.left:
                return _search(start.left, target)
            return None
        else:
            # right sub-tree
            if start.right:
                return _search(start.right, target)
            return None
```

```python
t1 = BST()
t1.insert(5)
t1.insert(3)
t1.insert(4)
t1.insert(1)
t1.insert(7)
t1.insert(9)
t1.insert(6)
t1.printBF()
t1.printDF(order="pre")
t1.printDF(order="post")
t1.printDF(order="in")
```
