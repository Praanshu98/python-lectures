### Beautiful Soup

```python
from bs4 import BeautifulSoup

url = "https://www.flipkart.com"
response = requests.get(url)
print(response.status_code)
content = response.content
soup = BeautifulSoup(content, 'html5lib')
print(soup)
imgList = soup.findAll('img') # find all the img tags
print(imgList[0]) # print one of the tags and understand it's details
imgList = soup.findAll('img', {'class': '_1Nyybr _30XEf0'})
img = imgList[0]
print(img['src'])
```

### URLs and Drawback

Sometimes the source may contain the relative URL, and in the case, we need to add the base URL.

```python
for i in range(len(imgList)):
    imgTag = imgList[i]
    src = imgTag['src']
    imgRes = requests.get(src)
    with open(f'output/flipkart-{i}.jpg', 'wb') as file:
        file.write(imgRes.content)
```

If we paste the content that we get from Flipkart, we will notice that we are not getting the full page and this is one of the drawbacks of web scrapping.
