### Style and Labelling

```python
x = list(range(1, 100, 3))
y = [t**2 for t in x] # x^2
plt.plot(x, y, "g-*", markersize=5)
```

```python
import numpy as np
x = np.arange(1, 102, 5)
y = x**2 # numpy is a vectorized library

plt.title("Square")
plt.xlabel("x-values")
plt.ylabel("y-values")
plt.xlim(1, 101)
plt.ylim(1, 10000)
plt.plot(x, y, 'r-*', markersize=10)
```

### Plotting multiple graphs

```python
import numpy as np
x = np.arange(1, 10)
y_square = x**2
y_cube = x**3

plt.title("Square vs Cube")
plt.plot(x, y_square, label="square")
plt.plot(x, y_cube, label="cube")
plt.legend()
plt.xlabel("x-values")
plt.ylabel("y-values")
plt.show()
```

```python
import numpy as np
x = np.arange(1, 10)
y_square = x**2
y_plus_beta = x + 50

plt.plot(x, y_square,"y-*", label="square")
plt.plot(x, y_plus_beta,"g--o", label="x+50")
plt.legend()
plt.xlabel("x-values")
plt.ylabel("y-values")
plt.text(2.7, 57, "(3, 53)")
plt.show()
```

```python
import math
x = np.arange(1, 101, 5)
# y = [(i+50)**2/2 for i in x]

def custom_func(x):
    return (x + 50)**2 / 2

vectorized_log = np.vectorize(custom_func) 
# converts the function into a vectorized function

y = vectorized_log(x)
plt.plot(x, y, '-o')
plt.show()
```

### Bar Graphs

```python
plt.bar([1, 2, 3], [10, 20, 5])
plt.show()
```

```python
width = 0.2
x = np.array([1, 2, 3])
y_1 = np.array([10, 20, 6])
y_2 = np.array([5, 10, 5])
plt.bar(x-(width/2), y_1, width=width, color="b", label="crime rate")
plt.bar(x+(width/2), y_2, width=width, color="r", label="murder rate")
plt.legend()
plt.show()
```

### Histogram - how values are distributed

```python
vals = [1, 2, 3, 1, 2, 1, 4, 5, 2, 1, 4, 6, 3, 2, 2, 2]
plt.hist(vals)
plt.show()
```

```python
vals = np.random.rand(100)

plt.hist(vals, [0, 0.4, 0.7, 1])
plt.show()
```

```python
mu = 1000
sigma = 15
vals = np.random.normal(mu, sigma, 100)
count, bins, _ = plt.hist(vals, bins=10)
plt.show()
print(count)
print(bins)
```

### Scatter Plot

```python
x = np.random.rand(100)
y = np.random.rand(100)

x1 = np.random.rand(100) + 0.5
y1 = np.random.rand(100)

plt.scatter(x, y)
plt.scatter(x1, y1)
plt.show()
```

### Plot a custom function

```python
def sigmoid(x):
    return math.tanh(x)

vectorized_sigmoid = np.vectorize(sigmoid)

x = np.arange(-10, 10, 0.1)
y = vectorized_sigmoid(x)

plt.plot(x, y)
plt.show()
```
